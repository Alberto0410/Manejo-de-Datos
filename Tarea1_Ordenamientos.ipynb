# Descripción de los algoritmos

##InsertionSort
Este algoritmo consiste en ir insertando cada uno de los elementos en su posición adecuada.

*Descripción:*
1. Comenzamos con un solo elemento (del conjunto a ordenar), por lo tanto asumimos que ese elemento ya se encuentra en la posición correcta.
2. Posteriormente ya que se tienen  k  elementos ordenados tomamos al elemento  k+1 , lo comparamos con todos los elementos ya ordenados y se identifica la posición en la que debe insertarse este elemento.
3. Se inserta el elemento  k+1  en la posición correspondiente y el resto de elementos se desplazan un lugar.
4. Una vez que se ha repetido este proceso con todos los elementos a ordenar, podemos garantizar que el conjunto ya ha sido ordenado.


## SelectionSort
Recorre toda la lista dada en busca del elemento menor o mayor, dependiendo del sentido en el que se quiera ordenar; una vez localizado se lleva al primer o último lugar, dependiendo del caso 
Este proceso lo repite hasta obtener la lista ordenada

*Descripción*

1. Se recorre la lista dada y se toma el elemento máximo o mínimo

2. Se coloca en su posición donde debería ir (ya sea el primero o el último lugar) y el resto de elementos se recorren  a la derecha si el orden es de mayor a menor, en caso contrario es a la izquierda

3. Regresa al paso 1 y empieza a recorrer la lista a partir del elemento 2 debido a que el primero ya está en su lugar correspondiente. Una vez encontrado el elemento mínimo se coloca a la derecha del primero. En caso de que el orden sea de mayor a menor se hace su analogo

4. Una vez terminas las iteraciones se devuelve la lista ordenada


##BubbleSort
Este algoritmo toma parejas de elementos del conjunto de izquierda a derecha y va colocando en la posición de la izquierda al valor más pequeño de ambos elementos, se hace esto con las parejas iniciales y al final de la primera pasada podemos asegurar que el elemento que queda en la posición más a la derecha de la colección es el elemento de mayor y ya se encuentra en su posición. Al repetir este proceso el suficiente número de veces, la colección queda ordenada.

*Descripción*:

1. Comenzamos la colección desordenada, y se toma la primer pareja de elementos a ordenar, se pasa a la casilla izquierda al menor.
2. Tomamos la siguiente pareja a la derecha (considerando al elemento que quedo en la casilla derecha de la pareja anterior).
3. Se compara nuevamente a ambos elementos y se coloca del el mayor de ambos en la casilla derecha.
4. Una vez que se itero por primera vez a todos los elementos de la colección, podemos asegurar que el elemento en la posición más a la derecha es el mayor y ya se encuentra en su posición final.
3. Se repiten los pasos 2, 3 y 4 hasta que todos los elementos estén ordenados y tomando en cuenta que en cada iteración se coloca en la posición que le corresponde al elemento más a la derecha, por lo tanto ya no hace falta comparar los elementos que ya hayan sido colocados en su posición final.

##MergeSort
Se tiene un conjunto o estructura de datos (lista, vector, arreglo, etc) de tamaño  n  que se encuentran desordenados y que mediante algún atributo podemos ordenar, la idea de este algoritmo es la siguiente:

*Descripción*:

1. Partir el conjunto en 2 subconjuntos de tamaño $\frac{n}2$.
2. Posteriormente volver a dividir estos en 4 subconjuntos de tamaño $\frac{n}4$, se repite este proceso hasta que se tienen subconjuntos de tamaño 2.
3. Ya que se tienen conjuntos de tamaño a lo mas 2, se ordenan de manera individual cada uno de ellos.
4. Cada uno de estos subconjuntos ordenados se mezclan de manera recursiva y ordenada, hasta que todo el conjunto de datos se encuentran ordenados.

##QuickSort

Se tiene un conjunto o estructura de datos (lista, vector, arreglo, etc) de tamaño  $n$  que se encuentran desordenados y que mediante algún atributo podemos ordenar. La diferencia entre ambos algoritmos es que QuickSort se toma un pivote para poder acomodar a los menores y mayores.

*Descripción*

1. Tomamos un valor de la lista de datos a ordenar (que idealmente parte a la colección a la mitad), a ese valor le llamamos pivote.
2.  Una vez que se eligió al pivote, comparamos a todos los elementos contra el pivote y almacenamos los menores al lado izquierdo del pivote y a los mayores del lado derecho los cual nos deja 2 subconjuntos de tamaño $\frac{n}2$.
3. Posteriormente volvemos a tomar un pivote en cada una de las respectivas subconjuntos, e idealmente volvemos a dividir estas en 4 subconjuntos de tamaño $\frac{n}4$, se repite este proceso hasta que se tienen subconjuntos de tamaño 2.
4. Ya que se tienen conjuntos de tamaño a lo mas 2, se ordenan de manera individual cada uno de ellos.
5. Cada uno de estos subconjuntos ordenados se mezclan de manera recursiva y ordenada, hasta que todo el conjunto de datos se encuentran ordenados.


#Análisis de los algorítmos

## SelectionSort

No hay peor caso, puesto que en cada una de las iteraciones se compara el mínimo auxiliar con todos los demás elementos hasta encontrar el mínimo de la lista y los intercambia. 

Cada vez que se encuentra al mínimo de la lista de tamaño $n$ y se coloca en su lugar, se repite el alogritmo para la lista que va desde el segundo elemento hasta el último, es decir, para una lista de tamaño $(n-1)$, por lo que para una lista de tamaño $n$ realizará el mismo número de pasos que para una lista de tamaño $(n-1)$ más el número de comparaciones que realizó para encontrar el mínimo, las cuales fueron $(n-1)$.

Por lo tanto, si $T(n)$ es el número de pasos que realiza para ordenar una lista de tamaño $n$

$$T(n)=T(n-1)+(n-1)$$

##InsertionSort
Dada la naturaleza del algoritmo, el peor caso en cuando la lista dada se encuentra ordenada en el sentido contrario en el cual queremos

Sabemos que en todos los casos el algoritmo va a recorrer y comparar todos los elementos, pero cuando está en la situación descrita arriba, entonces va a tener que hacer en cada iteración un intercabio, lo cual le cuesta una operación más.

Es decir, siempre que compare si el elemento de la derecha es menor que el de la izquierda, va a realizar el intercambio ya que está ordenada en sentido contrario y por lo tanto dicha condición siempre se cumple

## BubbleSort
El peor caso es cuando la lista de $n$ está ordenada en sentido contrario al que queremos, pues al tomar cualquier par de elementos consecutivos de la lista, siempre se deberá hacer un intercambio, además de que aunque se llegue a acomodar el último elemento en su lugar, la lista de $n-1$ elementos que queda para la segunda vuelta de comparaciones seguirá estando ordenada en sentido contrario, por lo que nuevamente cada que se tome un par consecutivo de elementos de dicha lista será necesario hacer un intercambio

## MergeSort
Sea $T(n)=$ número de pasos que realiza el algoritmo para una lista de n elementos.

Cuando la lista tiene 0 o 1 elementos, podemos decir que la lista ya se encuentra ordenada. Por lo que en este caso, hace una validación y entonces $T(n)=1$

Cuando $n \geq 2$ divide la lista a la mitad y realiza el algoritmo a cada una de las sublistas, por lo que a cada una de ellas, le tomará el doble de pasos que al  de la mitad de su lista principal y además el número de las operaciones que le lleva combinar las listas ($a\cdot n$), es decir $2T(\frac{n}{2})+a\cdot n$.

$$T(n)=\begin{cases}
1 & n=0\,\acute{o}\,n=1\\
2T(\frac{n}{2})+a\cdot n & n\geq2
\end{cases}$$

##QuickSort
El pero caso que puede existir en este algoritmo está sumamente relacionado con el pivote, ya que si éste es el elemento mínimo, entonces la lista de los número menores a él va a ser una lista vacia, mientras que la lista de los número mayores va a ser la lista complemento del pivote.

De esta manera, si hacemos que el pivote siempre sea el elemento mínimo de la lista, entonces en escencia lo que se está haciedo es elegir el número más chico y ponerlo en la posición incial, en otras palabras, estamos usando el algoritmo insertion y por lo visto en clase (y más abajo), es de complejidad $n^2$.\
Por lo tanto, en el peor de los casos, la complejidad del Quick es $n^2$

Por otro lado, hay que considerar la complejidad del mejor caso, el cual es cuando elije como pivote el elemento que debería ir en medio en la lista ordenada

En este caso supongamos que la lista tiene longuitud $n = 2^k$, entonces se crean dos listas de longuitud $\frac{n}{2}$ y por cada elemento de la lista hace $\frac{n}{2}$ comparaciones, es decir que en total hace $2 \frac{n}{2}= n $ comparaciones. Cuando vuelve a elegir al pivote como el elemento de enmedio entonces hay 4 cuatro listas de longuitud $\frac{n}{4}$ y en cada una hace $\frac{n}{4}$ comparaciones y al final hace $n$ operaciones

Así, tenemos que hace en total $\underbrace{n+n+n+\cdots +n}_{k\, veces} = k\cdot n = n\log_{2}(n)$
Por lo tanto, eligiendo el mejor pivote, el algoritmo se convierte de complejidad $n\log_{2}(n)$

## MergeSort
Como $$T(n)=\begin{cases}
1 & n=0\,\acute{o}\,n=1\\
2T(\frac{n}{2})+a\cdot n & n\geq2
\end{cases}$$

**P.D. $T(n) \in O(n\log n)$**

Para encontrar la cota, supongamos que $n = 2^k$ para algun $k \in \mathbb{N}$
\begin{align*}
T(n) &= 2T(\frac{2^k}{2}) + a\cdot 2^k\\
            &= 2T(2^{k-1})+ a\cdot 2^k\\
            &= 2[2T(\frac{2^{k-1}}{2})+a\cdot 2^{k-1}] + a \dot 2^k\\
            &= 2^2T(2^{k-2}) + a\cdot 2^k + a\cdot 2^k\\
            &= 2^2[2T(\frac{2^{k-2}}{2}) + a\cdot 2^{k-2}] + 2a \cdot 2^k\\
            &= 2^3[T(2^{k-3})] + 3a\cdot 2^k  
    \end{align*}
Cuando se realiza $i$ veces obtenemos que:

$T(n)=2^iT(2^{k-i})+i\cdot a\cdot 2^k$

Cuando $i=k$ obtenemos que:
    \begin{align*}
        T(n) &= 2^kT(2^{k-k})+k\cdot a\cdot 2^k~~~~~~~(Sustituyendo~el~valor~de~i)\\
             &= 2^kT(2^0)+k\cdot a\cdot 2^k~~~~~~~~~~(k-k=0)\\
             &=2^kT(1)+k\cdot a\cdot 2^k~~~~~~~~~~~~(2^0=1)\\
             &=2^k(1)+ k\cdot a \cdot 2^k~~~~~~~~~~~~~~~(T(1)=1)\\
             &=2^k+k\cdot a \cdot 2^k~~~~~~~~~~~~\\
             &=n+k\cdot a\cdot n~~~~~~~~~~~~~~~~~~~~~~~(2^k=n)\\
             &=n+ a\cdot  n \cdot \log_2 n ~~~~~~~~~~~~~~~(k=\log_2n)
    \end{align*}

$$\therefore T(n)=n+ a\cdot  n \cdot \log_2 n$$

Como $a$ es una constante positiva, podemos concluir que la función de $T(n)$ que crece más rápido es:
$$ a\cdot  n \cdot \log_2 n$$


Por lo que esta sería la función que acota superiormente al número de pasos del algoritmo *MergeSort*.

Por último, tomando $c=a$,podemos concluir que:
 $$\therefore T(n)\in O(n\cdot\log_{2}n)\blacksquare$$

````python
# MHDA
# OTA
# Implementacion del algoritmo seleccion
def seleccion(lista):
  # asigno el largo de la lista en una variable:
  n= len(lista)
  
  # Con este for ordenamos cada lista de n-i elementos
  for i in range(0,n):
    # asignamos una variable para el minimo de la lista
    minimo=i
    # en este for encontraremos el valor minimo de la lista
    for j in range(i+1,n):
      if lista[j]<lista[minimo]:
        # reasignamos el valor del minimo
        minimo=j
    # intercambiamos de lugar al primer elemento con el minimo
    lista[i],lista[minimo]=lista[minimo],lista[i]
  # se devuelve la lista ordenada
  return lista

# MHDA
# OTA
#Algoritmo Insertion. Recibe como argumentos una lista desordenada
def insertar(lista):
  #Iteramos sobre cada elemento de la lista
  for elemento in range(len(lista)):
    #Iteramos sobre los numeros anteriores a "elemento"
    for anterior in range(elemento,0,-1):
      #Si hay un numero más chico que "elemento", entonces los intercambiamos
      if lista[anterior]<lista[anterior-1]:
        lista[anterior],lista[anterior-1]=lista[anterior-1],lista[anterior]
  #Regresamos la lista ordenada
  return lista

# MHDA
# OTA
def burbuja(lista):
  #Guardamos la longuitud de la lista y creamos una variable bandera
    n = len(lista)
    bandera = True
    #Incializamos el contador que va a correr por los indices de la lista
    j = 0
    while(bandera):
      #Cada iteración aumentamos el indice
        j = j+1
        #Cambiamos el estado de la bandera, nos avisa si ya terminó las comparaciones
        bandera = False
        #Bucle que va comparando por parejas
        for i in range(n-j):
          #Si el elemento de la izquiera es menor que el de la derecha entonces intercambia
            if lista[i] > lista[i+1] :
                lista[i], lista[i+1] = lista[i+1], lista[i]
                #Indicamos que sí hubo un intercambio
                bandera = True
    return lista

# MHDA
# OTA
# Implementacion del algoritmo unir
def unir(lista):
  # Si la longuitud es 0 o 1 entonces ya esta ordenada
  if len(lista)<2:
    return lista
  
  # Si no es así, entonces dividimos la lista a la mitad
  mitad=len(lista)//2
  # Asignamos una variable para cada mitad de la lista
  sublista1=lista[:mitad]
  sublista2=lista[mitad:]
  # Se aplica el algoritmo a cada una de las sublistas
  unir(sublista1)
  unir(sublista2)

  # Definimos 3 variables que nos ayudaran con los indices
  # de los elementos de las sublistas
  i,j, k = 0,0,0

  # El proceso se repetira hasta que hayamos pasado por
  # todos los elementos de ambas sublistas
  while i < len(sublista1) and j< len(sublista2):
    # Comparamos los elementos de ambas listas que se
    # ecuentran en la misma posicion
    if sublista1[i] < sublista2[j]:
      # Se agrega el elemento minimo a la lista principal
      lista[k]=sublista1[i]
      # Se aumentan en una unidad las variables de apoyo
      i+=1
      k +=1
    else:
      # Se agrega el elemento minimo a la lista principal 
      lista[k]=sublista2[j]
      # Se aumentan en una unidad las variables de apoyo
      j+=1
      k +=1
  
  # Cuando la sublista 1 es mayor que la segunda, se 
  # agregan los elementos sobrantes a la lista principal
  while i<len(sublista1):
    lista[k]=sublista1[i]
    # Se aumentan en una unidad las variables de apoyo
    i += 1
    k += 1

  # Cuando la sublista 2 es mayor que la segunda, se 
  # agregan los elementos sobrantes a la lista principal
  while j<len(sublista2):
    lista[k]=sublista2[j]
    # Se aumentan en una unidad las variables de apoyo
    j += 1
    k += 1
  
  return lista


# MHDA
# OTA
#Funcion que separa los numeros con base en el pivote del QuickSort
def crear_sublistas(lista):
  #Creamos las listas para almacenar los elementos
  izquierda = []
  derecha = []
  #Elegimos el pivote como el elemento incial de la lista
  pivote = lista[0]
  
  #Iteramos desde el elemento 2 de la lista, pues el primero es el pivote
  for i in range(1, len(lista)):
    #Si el elemento i es menor que el pivote, va a la lista izquierda
    if lista[i] < pivote:
      izquierda.append(lista[i])

    #En caso contrario va a la lista derecha
    else:
      derecha.append(lista[i])
  #Regresamos el pivote y las sublistas generadas
  return izquierda, derecha, pivote


# MHDA
# OTA
#Algoritmo rapido que recibe como entrada una lista
def rapido(lista):
  #Si la longuitud es 0 o 1 entonces ya esta ordenada
  if 2>len(lista):
    #Regresamos la lista
    return lista
  #Si no es así, entonces llamamos a la función crear_sublistas
  else:
    izquierda, derecha, pivote = crear_sublistas(lista)
    #Hacemos recursion llamando quickSort para cada sublista generada
    return rapido(izquierda) + [pivote] + rapido(derecha)
  
````
