###Insertion

$$\left[15,e^{1},0,10,\sqrt{2},4,\pi,-1\right]$$

$$\left[e^{1},15 ,0,10,\sqrt{2},4,\pi,-1\right]$$

Considerando la clase `AlumnoFC` vista en clase, programa el ordenamiento *MergeSort* y prueba tu algoritmo con un conjunto de objetos de tipo `AlumnosFC`. Así como se muestra en la [presentación](https://colab.research.google.com/github/jugernaut/ManejoDatos/blob/main/AlgoritmosBusqueda/02_BusquedaBinaria.ipynb).
```python
#Copiamos la clase AlumnoFC
class AlumnoFC(object):

  def __init__(self, nombre, numero_cuenta, calif):
    self.nombre = nombre
    self.numero_cuenta = numero_cuenta
    self.calificaciones = calif

  def __eq__(self, otro):
    if self.numero_cuenta == otro.numero_cuenta:
      return True
    else:
      return False

  def __ge__(self, otro):
    if self.numero_cuenta >= otro.numero_cuenta:
      return True
    else:
      return False

  def __le__(self, otro):
    if self.numero_cuenta <= otro.numero_cuenta:
      return True
    else:
      return False

  def __gt__(self, otro):
    if self.numero_cuenta > otro.numero_cuenta:
      return True
    else:
      return False

  def __lt__(self, otro):
    if self.numero_cuenta < otro.numero_cuenta:
      return True
    else:
      return False

  def __str__(self):
    return self.nombre
  
  def promedio(self):
    promedio = sum(self.calificaciones)/len(self.calificaciones)
    return promedio
```

```python
#Escribimos el algoritmo de Merge(Unir)
def unir(lista:list) -> list:

  '''
  Funcion que realiza el algoritmo de Mergesort
  Entrada:
        lista: coleccion de datos
  Salida: la colección de datos ordenada en sentido creciente
  >>>unir([7,2,10,-2,1])
  [-2, 1, 2, 7, 10]
  '''
  #Si la lista tiene longitud 1, ya está ordenada
  if len(lista)<2:
    return lista
  
  #Partimos en dos la lista
  mitad=len(lista)//2

  #Creamos las sublistas a partir del elemento "mitad"
  izquierda=lista[:mitad]
  derecha=lista[mitad:]
  #Llamamos al algoritmo unir con cada sublista
  unir(izquierda)
  unir(derecha)

  #Creamos tres contadores
  #Dos son para recorrer las dos sublistas
  #El tercero es para decirnos que ya recorrio toda la lista
  i,j, k = 0,0,0

  #Mientrar los indices de cada lista no hayan superado 
  #la longitud de su respectiva lista...
  while i < len(izquierda) and j< len(derecha):
    #Si el elemento de la izquierda es menor que el otro:
    if izquierda[i] < derecha[j]:
      #Ponlo en su lugar
      lista[k]=izquierda[i]
      i+=1
      k +=1
    #Si ocurre lo contrario
    else:
      lista[k]=derecha[j]
      j+=1
      k +=1
  
  #Estos dos whiles sirven para recorrer los elementos faltantes
  #en caso de que alguna lista haya tenido más elementos que otra
  while i<len(izquierda):
    lista[k]=izquierda[i]
    i += 1
    k += 1

  while j<len(derecha):
    lista[k]=derecha[j]
    j += 1
    k += 1
  
  return lista
```

```python
Mike = AlumnoFC('Mike', 30134563, [10, 10, 10, 9, 10])
Aitor = AlumnoFC('Aitor', 20134563, [8, 10, 9, 9, 10])
Pete = AlumnoFC('Peter', 10134563, [9, 8, 8, 9, 10])

#Creamos la lista de alumnos
alumnos = [Mike, Aitor, Pete]

print("Alumnos desordenados:")
for alumno in alumnos:
  print("*", alumno)

#Los ordenamos
alumnosOrdenados = unir(alumnos)

print("\nAlumnos Ordenados: ")
for alumno in alumnosOrdenados:
  print("**", alumno)
```

## (1.5 Puntos) Orden de complejidad

Demostrar de **manera formal** el orden de complejidad al que pertenece el algoritmo de ordenamiento *QuickSort*, respecto a las operaciones (tiempo). También, explique a que orden de complejidad pertenece respecto a la memoria, considerando la versión recursiva.

### Procedimiento:

Al hacer el análisis del algoritmo *QuickSort* vamos a considerar el peor caso, esto es, cuando el elemento pivote es el mínimo. Al hacer esto, en realidad estamos seleccionando el elemento más chico y lo colocamos en su lugar, en otras palabras, estamos aplicando el algoritmo *Insertion*, pues lo que hace es buscar entre todos los elementos el más chico, llevarlo a su lugar y repetir el procedimiento pero ahora desde el segundo lugar

Dado que el algoritmo *Insertion* pertenece a $O(n^2)$, entonces, en el peor caso, el algoritmo *QuickSort* también pertenece a $O(n^2)$

Por otro lado, consideremos un caso promedio, es decir cualquier otro que no sea el peor. Lo que hace *Quick* es tomar un pivote y crear dos sublistas, una de ellas con todos los elementos menores que el pivote y otra con los mayores. Una vez hecho esto se le aplica el algoritmo *Quick* a cada una de las sublistas generadas

Dicho esto, consideremos sin pérdida de la generalidad una lista de tamaño $n = 2^k$. El primer paso es tomar el pivote y hacer $n$ operaciones, pues el algoritmo va a recorrer toda la lista seleccionando los menores y los mayores que el pivote. Una vez hecho esto, tenemos dos listas de $2^{k-1}$ a las cuales les vamos a aplicar el mismo proceso, es decir, vamos a tomar dos pivotes y se van a hacer $2^{k-1}$ comparaciones por cada sublista, esto es:
$$2^{k-1} + 2^{k-1} = 2(2^{k-1}) = 2^k$$
Luego, obtenemos 4 sublistas y se realizaran $\dfrac{2^{k-1}}{2} = 2^{k-2}$ comparaciones por cada una de ellas, es decir:
$$2^{k-2} + 2^{k-2} + 2^{k-2} + 2^{k-2} = 4(2^{k-2}) = 2^2(2^{k-2}) = 2^k$$
Así, notamos que por cada paso se hacen $2^k$ operaciones

Ahora la pregunta es, ¿cuál es el número de pasos que se van a hacer?
El proceso se llevara acabo $k$ veces, pues cada vez estamos diviendo entre dos la lista y el paso base es cuando la longitud de la lista es 1. De esta manera, el número de operaciones es:
$$T(n) =k \cdot 2^{k} = k\cdot n$$
Pero como habiamos dicho que $n  = 2^k$, entonces $k = \log_{2}(n)$
Por lo tanto:
$$ T(n) = n \cdot \log_{2} (n) $$

Finalmente, tenemos que dada la definición de cota superior asintótica se concluye que 
$$T(n) \in O(n\cdot \log_{2}(n) $$


Por otro lado, analizando el manejo de memoria, supongamos que tenemos una lista de $n = 2^k$ elementos.
La primera llamada se hace para la lista de $n$ elementos, depues se hacen dos llamadas, cada una para una lista de longitud $2^{k-1}$ pues dividimos a la lista inicial en dos de igual tamaño. Luego se hacen 4 llamadas para listas con $2^{k-2}$ elementos. El siguiente paso es hacer 8 llamadas porque tenemos 8 listas de tamaño $2^{k-3}$

Notamos que el número de llamadas está relacionado con la suma de potencias de dos, es decir que:
$$M(n) = \sum_{i =0}^{k} 2^{i} $$
Usando formulas de la series geométricas y propiedades de logaritmos:
$$M(n) = \frac{1-2^{k+1}}{1-2} =2^{k+1} -1  = 2^{\log_{2}(n) +1}-1 = 2^{\log_{2}(n) +\log_{2}(2)}-1 $$

$$M(n) = 2^{\log_{2}(n\cdot2)}-1 =2n -1 $$
Por lo tanto $M(n) = 2n - 1$

Acotando la función:

$$M(n) = 2n -1 < 2n $$
Por lo tanto, el orden de complejidad respecto al manejo de memoria del algoritmo Merge es $O(n)$

```python
#Creamos la clase Alumno 
class Alumno():
  #Definimos el constructor con los atributos requeridos
  def __init__ (self, correo, telefono):
    self.correo = correo
    self.telefono = telefono

   #Sobrecargamos el operador __str__
  def __str__ (self):
     return "Correo: " + self.correo +" Teléfono: " + str(self.telefono)

  #Sobrecargamos el operador igual
  def __eq__(self, otro):
    if self.correo == otro.correo:
      return True
    else:
      return False

  #Sobrecargamos el operador menor o igual
  def __ge__(self, otro):
    if self.correo >= otro.correo:
      return True
    else:
      return False

  #Sobrecargamos el operador mayor o igual
  def __le__(self, otro):
    if self.correo <= otro.correo:
      return True
    else:
      return False

  #Sobrecargamos el operador mayor que
  def __gt__(self, otro):
    if self.correo > otro.correo:
      return True
    else:
      return False

  #Sobrecargamos el operador menor que
  def __lt__(self, otro):
    if self.correo < otro.correo:
      return True
    else:
      return False
```

```python
#Llamos a los algoritmos iterativos de busqueda
#esto lo hacemos para no hacer tantas llamadas recursivas
#y así obtener un menor uso de la memoria
def busqueda_secuencialI(lista:list, elemento:int) -> int:
  '''
  Funcion que realiza el algoritmo de busqueda secuencial de manera iterativa
  Entrada:
        lista: coleccion de datos
        elemento: elemento a buscar en la lista(no necesariamente int)
  Salida: El indice si el elemento se encuentra en la lista
          False si el elemento no se encuentra en la lista

  >>>>>>busqueda_secuencialI([0,2,4,10,7], 4)
  True
  '''
  for i in range(len(lista)): #Recorremos toda la lista
    if lista[i] == elemento: #En cada iteración validamos si el elemento es el buscado
      return True #Si es así, lo retornamos
  return False #En otro caso regresamso un False

def busqueda_binariaI(lista:list, elemento) -> int:
  '''
  Funcion que realiza el algoritmo de busqueda de manera iterativa
  Entrada:
        lista: coleccion de datos ordenada
        elemento: elemento a buscar en la lista(no necesariamente int)
  Salida: El indice si el elemento se encuentra en la lista
          False si el elemento no se encuentra en la lista

  >>>busqueda_binariaI([0,2,4,10,7], 4)
  True
  '''

  inicio = 0 # Esta variable indica el primer indice de la lista
  final = len(lista)-1 # Esta variable indica el ultimo indice de la lista
  medio = 0 # Declaramos la variable que se usara en el ciclo while
  bandera = True # Variable de ayuda para la continuacion del ciclo while
  while bandera:
    if inicio > final: # Se evaula si la lista tiene longitud 0
      return False
    medio = (inicio + final) // 2 # Se guarda en una variable el indice del 
                                  # elemento que se encuentra a la mitad de la lista
    if lista[medio] < elemento: # Se evalua si el elemento de el medio se encuentra
                                # en el intervalo izquierdo de la coleccion
      inicio = medio + 1 # Se le da valor al primer indice de la lista
                         # de longitud n-2
    elif elemento < lista[medio]:# Se evalua si el elemento de el medio se encuentra
                                # en el intervalo derecho de la coleccion
      final = medio - 1 # Se le da valor al ultimo indice de la lista
                         # de longitud n-2
    if lista[medio] == elemento: # Se evalua si el elemento que se encuentra en
                                 # el medio es el elemento buscado
      return True
```

```python
#primero traemos los algoritmos de la tarea 1
def seleccion(lista) -> list:
  '''
  Funcion que realiza el algoritmo de Selectionsort
  Entrada:
        lista: coleccion de datos
  Salida: la colección de datos ordenada en sentido creciente
  >>>unir([7,2,10,-2,1])
  [-2, 1, 2, 7, 10]
  '''
  # asigno el largo de la lista en una variable:
  n= len(lista)
  
  # Con este for ordenamos cada lista de n-i elementos
  for i in range(0,n):
    # asignamos una variable para el minimo de la lista
    minimo=i
    # en este for encontraremos el valor minimo de la lista
    for j in range(i+1,n):
      if lista[j]<lista[minimo]:
        # reasignamos el valor del minimo
        minimo=j
    # intercambiamos de lugar al primer elemento con el minimo
    lista[i],lista[minimo]=lista[minimo],lista[i]
  # se devuelve la lista ordenada
  return lista

def insertar(lista) -> list:
  '''
  Funcion que realiza el algoritmo de Insertionsort
  Entrada:
        lista: coleccion de datos
  Salida: la colección de datos ordenada en sentido creciente
  >>>unir([7,2,10,-2,1])
  [-2, 1, 2, 7, 10]
  '''
  #Iteramos sobre cada elemento de la lista
  for elemento in range(len(lista)):
    #Iteramos sobre los numeros anteriores a "elemento"
    for anterior in range(elemento,0,-1):
      #Si hay un numero más chico que "elemento", entonces los intercambiamos
      if lista[anterior]<lista[anterior-1]:
        lista[anterior],lista[anterior-1]=lista[anterior-1],lista[anterior]
  #Regresamos la lista ordenada
  return lista

def burbuja(lista):
    '''
  Funcion que realiza el algoritmo de Insertionsort
  Entrada:
        lista: coleccion de datos
  Salida: la colección de datos ordenada en sentido creciente
  >>>unir([7,2,10,-2,1])
  [-2, 1, 2, 7, 10]
  '''  
  #Guardamos la longuitud de la lista y creamos una variable bandera
    n = len(lista)
    bandera = True
    #Incializamos el contador que va a correr por los indices de la lista
    j = 0
    while(bandera):
      #Cada iteración aumentamos el indice
        j = j+1
        #Cambiamos el estado de la bandera, nos avisa si ya terminó las comparaciones
        bandera = False
        #Bucle que va comparando por parejas
        for i in range(n-j):
          #Si el elemento de la izquiera es menor que el de la derecha entonces intercambia
            if lista[i] > lista[i+1] :
                lista[i], lista[i+1] = lista[i+1], lista[i]
                #Indicamos que sí hubo un intercambio
                bandera = True
    return lista

def crear_sublistas(lista):
  '''
  Funcion que divide una lista dada en dos con respecto a un pivote
  Entrada:
        lista: colección de datos
  Salida: 
        izquierda: lista de elementos menores respecto al pivote
        derecha: lista de elementos mayores o igual al pivote
        pivote: elemento con respecto a quien se hace la comparacion
  >>>crear_sublistas([7,2,10,-2,1])
  ([2, -2, 1], [10], 7)
  '''
  #Creamos las listas para almacenar los elementos
  izquierda = []
  derecha = []
  #Elegimos el pivote como el elemento incial de la lista
  pivote = lista[0]
  
  #Iteramos desde el elemento 2 de la lista, pues el primero es el pivote
  for i in range(1, len(lista)):
    #Si el elemento i es menor que el pivote, va a la lista izquierda
    if lista[i] < pivote:
      izquierda.append(lista[i])

    #En caso contrario va a la lista derecha
    else:
      derecha.append(lista[i])
  #Regresamos el pivote y las sublistas generadas
  return izquierda, derecha, pivote


def rapido(lista):
  '''
  Funcion que realiza el algoritmo de Quicksort
  Entrada:
        lista: coleccion de datos
  Salida: la colección de datos ordenada en sentido creciente
  >>>unir([7,2,10,-2,1])
  [-2, 1, 2, 7, 10]
  '''
  #Si la longuitud es 0 o 1 entonces ya esta ordenada
  if 2>len(lista):
    #Regresamos la lista
    return lista
  #Si no es así, entonces llamamos a la función crear_sublistas
  else:
    izquierda, derecha, pivote = crear_sublistas(lista)
    #Hacemos recursion llamando quickSort para cada sublista generada
    return rapido(izquierda) + [pivote] + rapido(derecha)
  
```
