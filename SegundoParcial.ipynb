###Árboles binarios
Un árbol binario es una gráfica simple (no se permiten los lazos ni los grafos dirigidos), conexa y sin ciclos, la cual además sólo puede tener como máximo dos hijos.

*Ordenes de complejidad*
1. Inserta. El peor caso es cuando tenemos un árbol completo ya que siempre va a tener que recorrer por lo menos $n + 1$ nodos, con $n$ la altura del árbol.

  Consideremos que tenemos un árbol completo de altura n. Si queremos insertar un nodo entonces primero hacemos una operación que es llamar al método *insertar_nodo*; luego, como la raíz no es vacía, entonces va a elegir al azar si irse al hijo derecho o al izquierdo. Sin pérdida de la generalidad supongamos que siempre escoge el hijo izquierdo (esto se puede hacer ya que es irrelevante qué hijo escoja, siempre va a hacer el mismo número de pasos ya que estamos en un árbol completo). Así, va a llamar $n$ veces al método *insertar_nodo* porque va a ir recorriendo los nodos $n$ nodos de la izquierda hasta llegar a una hoja e insertar el nodo ahí. De esta manera, en total entró $n+1$ veces, pues son $n$ por cada nodo y $1$ por la raíz; además, en cada paso hace 1 validación y 3 asignaciones, por lo que:
  $$I(n) = 4(n+1) = 4n + 4$$
  Acotando la función, tenemos que: $I(n) = 4n + 4 \leq 4n + 4n = 8n $
  Se concluye que $I(n) \in O(n)$, es decir que su orden de complejidad es lineal

2. Busca. El peor caso es cuando tenemos un árbol completo, pues vamos a tener que recorrer a lo más $n$ nodos, pues consideramos que vamos a ir hasta alguna hoja. De esta menarea se tiene que como le método *busca* depende de la altura, entonces es de orden lineal, por lo que $B(n) \in O(n)$ 

3. Elimina. Dado que para eliminar un árbol completo basta con hacer la raíz igual a *None*, entonces el orden de complejidad es constante, pues sólo se necesita hacer una operación
  $$E(n) \in O(c)$$

4. Borra nodo. La clase Árbol binario no tiene definido éste método

###Árbol binario de búsqueda
Un árbol binario de búsqueda es un árbol binario en el cual se cumple que:

1. El hijo derecho es vacío o todos los elementos del hijo derecho son mayores que la raíz y además el hijo derecho es un árbol binario de búsqueda

2. El hijo izquierdo es vacío o tododos los elementos del hijo izquierdo son menores que la raíz y el hijo izquierdo es un árbol binario de búsqueda


1. Busca. El peor caso es cuando queremos buscar el elemento máximo en un árbol degenerado, lo que es equivalente a tener una lista ordenada de menor a mayor. El método *busqueda* va comparando cada vez si el nodo donde está parado es mayor o menor al valor que esta buscando y termina cuando llega al valor indicado. En el caso ya mencionado, el método *búsqueda* se convierte en el algoritmo de *búsqueda secuencial* y por lo visto el parcial pasado se concluye que el método tiene un orden de complejidad constante, es decir:
$$B(N) \in O(n)$$

2. Inserta. Analogamente al método *busca*, tenemos que el peor caso es cuando queremos insertar un elemento mayor a todos un árbol degenerado. Primero tenemos que buscar el nodo para verificar si existe o no, por lo que ya tenemos $n +1$ operaciones para buscar el nodo. Luego, al insertarlo tenemos que entrar $n+1$ veces al método *inserta_nodo* pues lo hacemos una vez por cada nodo. Dado que cada vez que se entra al método *inserta_nodo* se hacen 4 operaciones, entonces:
$$I(n) = n +1 + 4(n+1) = n + 1 + 4n + 4 = 5n + 5$$
Finalmente, acotando la función tenemos que $I(n) \leq 10n$, por lo que:
$$I(n) \in O(n)$$

3. Elimina. Basta con eliminar la raíz para poder eliminar todo el nodo, por lo que sólo tenemos que asignarle a la raíz *None*, es decir que se hace una única operación independientemente el caso. Entonces se conlcuye que:
$$E(n) \in O(c)$$


4. Borra nodo. El peor caso es cuando queremos eliminar el nodo con valor máximo en un árbol degenerado, pues tenemos que recorrer todos lo elementos para encontrarlo. Supongamos que tenemos un árbol degenerado de altura $n$ y queremos eliminar el elemento máximo. Vamos a tomar $n+1$ operaciones para la busqueda y para eliminarlo unicamente tenemos que hacer una operacion, la cual es asignar "nodo_a_borrar.padre.der = None", es decir, cortar la conexión con su padre. Entonces:
$$B(n) = n  + 1 + 1 = n + 2 \leq n + 2n = 3n$$
Por lo tanto, el orden de complejidad de *borrar_nodo* es lineal, es decir:
$$B(n) \in O(n)$$

La diferencia entre el AB y el ABB es que en el AB no hay un "orden", es decir que en el AB puede suceder que el hijo izquierdo y el derecho sean menores que su padre y no hay ningun problema, mientras que esto no puede pasar en los ABB

###Árbol binario de búsqueda balanceado
Un árbol de búsqueda balanceado es unárbol de búsqueda que cumple que:
1. Sus dos hijos son árboles de búsqueda balanceados
2. El valor absoluto de la diferencia entre las alturas de sus hijos es menor o igual que unos, es decir que:

$$|altura_{H_d} - altura_{H_i}| \leq 1$$

1. Busca. En realidad no hay pero caso, pues en todos los casos el árbol está distribuido de manera uniforme. Entonces, tomando cualquier AAVL, busquemos cualquier elemento. Notemos que cada vez que avanzamos hacia algun nodo estamos dejando atras la mitad de los elementos, pues estamos en un arbol balanceado, lo que nos indica que la altura del árbol izquierdo y el derecho es casi la misma. Así, notamos que en realidad estamos en el algoritmo de búsqueda binaria, porque primero nos preguntamos si donde estamos parados es mayor o menor que el elementos a buscar, despues acortamso la lista a la mitad y volvemos a hacer el procedimiento. De esta manera, por el periodo pasado tenemos que $B(n) \in O(log_2(n))$ 
2. Inserta. Con el mismo razonamineto que el inciso anterior, no hay peor caso. Ahora, primero tenemos que buscar el elemento, por lo que nos tomará $log_2(n)$ operaciones para buscarlo. Luego, como estamos trabajando con un AAVL, entonces hay dos casos, si el nodo en donde estamos es vacío, entonces haremos $a$ operaciones, con $a$ una constante (la cual indica el número de operaciones que le lleva balancear el árbol), si no es así, entonces en cada paso siguiente que damos estamos eliminando $n/2$ elementos, pues la cantidad de nodos está distribuida de manera lo más uniformemente posible. Entonces:

  $$ T(n)= \begin{cases}
a & n=1\\
T(\frac{n}{2})+a & n>1
\end{cases}$$

  **Veamos que $T(n) \in O(\log_2 n)$**

 Sin pérdida de la generalidad, sea $n=2^k$

  \begin{align*}
    T(n)  &=T(2^k)\\
          &=T(\frac{2^k}{2})+a~~~~~Def. ~de~ T(2^k)\\
          &=T(2^{k-1})+a~~~~Leyes~ exponentes\\
          &=(T(2^{k-2})+a)+a~~~~~Def.~ de~ T(2^{k-1})\\
          &=T(2^{k-2})+2\cdot a~~~~~Asociatividad\\
          &=(T(2^{k-3})+a)+2\cdot a~~~~~Def.~ de~ T(2^{k-1})\\
          &=T(2^{k-3})+3\cdot a~~~~~Asociatividad\\
    i-veces & \vdots & \\
           &=T(2^{k-i})+i\cdot a\\
    i=k		& \vdots & \\
            &=T(2^{k-k})+k\cdot a\\
            &=T(2^0)+k\cdot a\\
            &=T(1)+k\cdot a~~~~Leyes~ exponentes\\
            &=a+k\cdot a~~~~Def.~T(1)\\
            &=a+ \log_2 n \cdot a~~~~k=\log_2 n\\
            &=a+ a\log_2 n 
  \end{align*}
Por lo tanto, tenemos que el número total de operaciones, considerando las que le tomo en hacer la busqueda es:
$I(n) = log_2(n)(1+a) + a$
Se concluye que el orden de complejidad es $log_2(n)$, es decir que:
$$I(n) \in O(log_2(n))$$
3. Elimina. Analogamente a los otros dos ejercicios anteriores, sólo se tiene que hacer una operaciones, por lo que $E(n) \in O(c)$
4. Borra nodo.El peor caso sería eliminar el elementoq que está hasta la derecha. Analogamente al inciso anterior, primero tenemos que buscar el nodo, por lo que nos tomará $log_2(n)$ operaciones encontrarlo, luego tenemos que borrarlo, como no tiene hijos (pues es el de hasta la derecha), entonces unicamente se hace una operación y ya lo borramos. Luego, hay que rebalancear el árbol, lo que nos toma un número constante de operaciones, digamos $a$. Así, tenemos que:
$$B(n) = log_2(n) + 1 + a $$
Por lo tanto, el orden de complejidad es $log_2(n)$, es decir que:

$$B(n) \in O(log_2(n))$$

```python
# Tomamos la clase Nodo
class Nodo(object):
  '''
  Constructor de nodos
  dato: es el contenido del nodo
  '''
  def __init__(self, dato):
    self.izq = None
    self.dato = dato
    self.der = None
    self.padre = None

  # Nos dice si el nodo tiene al menos un hijo    
  def tieneHijos(self):
    return self.izq is not None and self.der is not None
  
  # Nos dide si el nodo es un hijo izquierdo
  def esIzq(self):
    return self.padre.izq == self
  
  # Nos dice si el nodo es un hijo derecho
  def esDer(self):
    return self.padre.der == self
  
  # Imprime el contenido del nodo
  def __str__(self):
    if self.dato is None:
      pass
    else:
      return '{}'.format(self.dato)

  #Nuevo metodo
  def cuantos_hijos(self):
    '''
    Funcion que realiza el conteo de cuantos hijos tiene un NODO
    Entrada:
          lista: self
    Salida: numero de hijos que tiene self

    >>>>>>nodo.cuantos_hijos()
    3
    '''
    #caso cuando no tiene hijos
    if self.izq is None and self.der is None:
      return 0
    #casos cuando tiene solo un hijo
    elif self.izq is not None and self.der is None:
      return 1
    elif self.izq is None and self.der is not None:
      return 1
    #caso cuando tiene ambos hijos
    elif self.izq is not None and self.der is not None:
      return 2
    
```

```python
import random

''' Clase AB (ArbolBinario), genera la estructura de un arbol binario, asi como sus funciones.'''
class AB(object):

  ''' Consructor de arboles binarios'''
  def __init__(self, dato):
    self.raiz = Nodo(dato)
      
  ''' Elimina todo el arbol'''
  def elimina(self):
    self.raiz = None
      
  ''' Devuelve el nivel del nodo que se le pasa como parametro.
      La raiz se ubica en el nivel cero'''
  def nivel(self, nodo):
    if nodo is None:
      return -1
    else:
      return 1 + self.nivel(nodo.padre)

  ''' Devuelve la altura a partir del nodo que se le pasa como parametro,
      si el arbol es vacio la altura es cero, si no se le suma 1 al maximo
      de las alturas de sus hijos'''
  def altura(self, nodo):
    if nodo is None:
      return 0
    else:
      return 1 + max(self.altura(nodo.izq), self.altura(nodo.der))
      
  ''' Inserta recursivamente un dato en el arbol de manera recursiva
      nodo: es el nodo a partir del cual se quiere insertar
      dato: es el dato que se quiere insertar
  '''
  def insertar_nodo(self, nodo, dato):
    # si el nodo es vacio ahi se crea el nuevo nodo
    if nodo is None:
      nuevo_nodo = Nodo(dato)
      return nuevo_nodo
    # dado que en los AB no hay un orden el nuevo nodo se inserta donde sea
    if bool(random.getrandbits(1)):
      nuevo_nodo = self.insertar_nodo(nodo.izq, dato)
      nodo.izq = nuevo_nodo
      nuevo_nodo.padre = nodo
    else:
      nuevo_nodo = self.insertar_nodo(nodo.der, dato)
      nodo.der = nuevo_nodo
      nuevo_nodo.padre = nodo
    #nodo guarda toda la ruta de donde sera insertado el dato
    #hasta caer en el caso base, es por eso que se devuelve    
    return nodo
  
  ''' Inserta un nodo en el arbol, a partir de la raiz'''
  def insertar(self, dato):
    #Se inserta el dato desde la raiz
    self.insertar_nodo(self.raiz, dato)
      
  ''' Recorre el arbol enorden e imprime cada uno de sus nodos'''
  def recorrido_enorden(self, nodo):
    if nodo is not None:
      self.recorrido_enorden(nodo.izq)
      print(nodo.dato)
      self.recorrido_enorden(nodo.der)

  ''' Recorre el arbol preorden e imprime cada uno de sus nodos'''
  def recorrido_preorden(self, nodo):
    if nodo is not None:
      print(nodo.dato)
      self.recorrido_preorden(nodo.izq)
      self.recorrido_preorden(nodo.der)

  ''' Recorre el arbol postorden e imprime cada uno de sus nodos'''
  def recorrido_postorden(self, nodo):
    if nodo is not None:
      self.recorrido_postorden(nodo.izq)
      self.recorrido_postorden(nodo.der)
      print(nodo.dato)
```

```python
''' Clase arbol binario de busqueda, genera la estructura de un arbol binario
    de busqueda asi como sus funciones.'''
class ABB(AB):
  ''' Consructor de arboles'''
  def __init__(self, dato):
    # llamamos al constructor de la clase padre con la palabra super
    super(ABB, self).__init__(dato)
    # y se agrega la variable ultimo agregado que sirve para clases posteriores
    self.ultimo_agregado = self.raiz

  ''' Busca un dato en el arbol y devuelve el nodo'''
  def busqueda(self, nodo, dato):
    # si la raiz es None o el dato esta contenido en el nodo,
    # se devuelve el nodo.
    if nodo is None or nodo.dato == dato:
      return nodo
    # si el dato es mayo entonces se busca del lado derecho
    if nodo.dato < dato:
      return self.busqueda(nodo.der, dato)
    # si no, se busca en el lado izquierdo
    else:
      return self.busqueda(nodo.izq, dato)

  ''' Inserta un elemento en el arbol a partir de un nodo'''
  def insertar_nodo(self, nodo, dato):
    # si el nodo es vacio ahi se crea el nuevo nodo
    if nodo is None:
      nuevo_nodo = Nodo(dato)
      self.ultimo_agregado = nuevo_nodo
      return nuevo_nodo
    # si el dato es menor que su padre, se inserta en el lado izquierdo
    if dato < nodo.dato:
      nuevo_nodo = self.insertar_nodo(nodo.izq, dato)
      nodo.izq = nuevo_nodo
      nuevo_nodo.padre = nodo
    # de no ser asi se inserta del lado derecho
    else:
      nuevo_nodo = self.insertar_nodo(nodo.der, dato)
      nodo.der = nuevo_nodo
      nuevo_nodo.padre = nodo
    # nodo guarda toda la ruta de donde sera insertado el dato
    # hasta caer en el caso base, es por eso que se devuelve    
    return nodo
  
  ''' Inserta un nodo en el arbol, a partir de la raiz'''
  def insertar(self, dato):
    # si el dato a insertar no se encuentra en el ABB se inserta
    if self.busqueda(self.raiz, dato) is None:
      # Se inserta el dato desde la raiz
      self.insertar_nodo(self.raiz, dato)

  ''' Busca el minimo en el subarbol a partir de nodo, este metodo sirve
      para poder eliminar nodos del arbol y mantener el orden'''
  def minimo_en_subarbol(self, nodo):
    # caso base en el que no hay minimo
    if nodo is None:
      return nodo
    # caso base en el que el nodo es el minimo ya que no tiene hijo izq
    if nodo.izq is None:
      return nodo
    # llamada recursiva para movernos al siguiente minimo
    return self.minimo_en_subarbol(nodo.izq)

  ''' Borra un nodo en el arbol, busca al nodo que contiene a dato
      y en caso de existir lo borra, se pueden dar multiples casos'''
  def borra_nodo(self, nodo, dato):
    # Caso 0) revisamos si el arbol es vacio.
    if self.raiz is None:
      return None
    # buscamos el nodo a borrar
    nodo_a_borrar = self.busqueda(nodo, dato)
    aux = nodo_a_borrar

    # Caso 0.1) si el dato no se encontro en el arbol no se puede borrar
    if nodo_a_borrar is None:
      return None

    # Caso 1) si el nodo a borrar es la RAIZ
    if nodo_a_borrar is self.raiz:
      # caso1.1) no tiene hijos, solo se borra la raiz
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is None:
        self.raiz = None
        self.ultimo_agregado = None
        return None

      # Caso1.2) solo se tiene hijo derecho, entonces se sube al hijo derecho
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is not None:
        self.raiz = nodo_a_borrar.der
        self.raiz.padre = None
        return self.raiz

      # Caso1.3) solo se tiene hijo izquierdo, entonces se sube al hijo izquierdo
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is None:
        self.raiz = nodo_a_borrar.izq
        self.raiz.padre = None
        return self.raiz

      # Caso1.4) tiene ambos hijos
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is not None:
        # buscamos el minimo en el subarbol derecho (minimo de los mayores)
        minimo = self.minimo_en_subarbol(nodo_a_borrar.der)
        aux = minimo.padre
        self.raiz.dato = minimo.dato
        self.borra_nodo(minimo, minimo.dato)
        return aux

    else: #Caso 2)
      # a partir de aqui se tienen 3 casos:
      # si no tiene hijos simplemente se borra el nodo
      # si tiene un solo hijo (ya sea izquierdo o derecho) se sube al unico hijo
      # tiene ambos hijos

      # es necesario identificar si el nodo a borrar es hijo izquierdo o derecho
      es_izquierdo = False
      if nodo_a_borrar.padre.izq == nodo_a_borrar:
        es_izquierdo = True

      # Caso2.1) no tiene hijos, solo se borra el nodo
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is None:
        aux = nodo_a_borrar.padre
        # revisamos si el nodo a borrar es un hijo izquiero o derecho
        if es_izquierdo: # Caso2.1.1)
          aux.izq = None
        else: # Caso2.1.2)
          aux.der = None
        nodo_a_borrar = None
        return aux
      
      # Caso2.2) solo se tiene hijo izquierdo, entonces se sube al hijo izquierdo
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is None:
        nodo_a_borrar.izq.padre = nodo_a_borrar.padre
        aux = nodo_a_borrar.padre
        # revisamos si el nodo a borrar es un hijo izquiero o derecho
        if es_izquierdo: # Caso2.2.1)
          nodo_a_borrar.padre.izq = nodo_a_borrar.izq
        else: # Caso2.2.2)
          nodo_a_borrar.padre.der = nodo_a_borrar.izq
        return aux

      # Caso2.3) solo se tiene hijo derecho, entonces se sube al hijo derecho
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is not None:
        nodo_a_borrar.der.padre = nodo_a_borrar.padre
        aux = nodo_a_borrar.padre
        # revisamos si el nodo a borrar es un hijo izquiero o derecho
        if es_izquierdo: # Caso2.3.1)
          nodo_a_borrar.padre.izq = nodo_a_borrar.der
        else: # Caso2.3.2)
          nodo_a_borrar.padre.der = nodo_a_borrar.der
        return aux

      # Caso2.4) tiene ambos hijos
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is not None:
        # buscamos el minimo en el subarbol derecho
        minimo = self.minimo_en_subarbol(nodo_a_borrar.der)
        aux = minimo.padre
        nodo_a_borrar.dato = minimo.dato
        self.borra_nodo(minimo, minimo.dato)
        return aux
  
  ''' Borra un nodo que contiene a dato a partir de la raiz'''
  def borrar(self, dato):
    self.borra_nodo(self.raiz, dato)

  def contar_hijos(self, dato):
    '''
    Funcion que regresa el número de hijos de un nodo de un ABB
    Entrada:
          lista: el ABB y el dato que tiene el nodo a buscar
    Salida: numero de hijos que tiene el nodo con el dato ingresado

    >>>>>>abb.contar_hijos(6)
    0
    '''
    nodo = self.busqueda(self.raiz, dato)
    return nodo.cuantos_hijos()
```

###Borra Nodo ABB
Básicamente hay dos casos, cuando el nodo a borrar tiene hijos o cuando no tiene hijos
1. Cuando en nodo a borrar no tiene hijos. En este caso sólo se tiene que hacer una operación, la cual es cortar su "conexión" con su padre, es decir que el hijo del padre del nodo a borrar va a ser None, que es lo mismo que *nodo_a_borrar.padre.izq = None* si es hijo izquierdo, o *nodo_a_borrar.padre.der = None* si es hijo derecho
2. Cuando tiene hijos. Aquí hay dos casos:

  2.1 Si tiene un hijo. Lo que se hace es subir el hijo al lugar del padre, lo anterior se logra asignando como nuevo padre del hijo del nodo a borrar al padre del nodo a borrar, es decir "nodo_a_borrar.izq.padre = nodo_a_borrar.padre* en caso de que el nodo a borrar tenga hijo izquierdo, si no es así, entonces se ahce su análogo si tiene hijo derecho.   2.2 Si tiene ambos hijos. En este caso se tiene que buscar el mínimo de los hijos mayores (aunque también se puede con el máximo de los hijos menores). Una vez encontrado dicho nodo se tiene que subir al lugar del nodo a borrar, esto se logra unicamente haciendo "nodo_a_borrar.dato = minimo.dato", lo cual actualiza el dato del nodo a borrar poniendo el dato del nodo mínimo. 2. Si el nodo está dos veces cargado a la derecha y se cumple que su hijo derecho tiene factor cero o factor 1, entonces hacemos una rotación simple a la izquierda
3. Si el nodo tiene factor -2 y su hijo izquierdo factor 1, entonces hacemos una rotación izquierda derecha
4. Finalmente, si el nodo tiene factor 2 y su hijo derecho factor -1, entonces hacemos una rotación derecha izquierda

```python
import re
from collections import Counter

class ContadorDePalabras():
  #Constructor con los tres atributos solicitados
  def __init__(self, ruta):
    self.ruta = ruta
    self.diccionario = {}
    self.diccionarioOrdenado = {}

  def machine_learning(self):
    #abrimos el archivo
    archivo = open(self.ruta).read().lower()
    #expresion regular que omite los espacios y signos de puntuacion
    expresion = '[a-zA-ZáéíóúÁÉÍÓÚñÑ]+'
    #aplicamos la expresion regular al texto
    lista_palabras = re.findall(expresion, archivo)
    #contamos todas la palabras
    for palabra in lista_palabras:
      self.diccionario[palabra] = lista_palabras.count(palabra)

  def ordena_diccionario(self):
    #usamos Counter para poder ordenar el diccionario
    diccionario_counter = Counter(self.diccionario)
    #lo asignamos al atributo diccionarioOrdenado
    self.diccionarioOrdenado = diccionario_counter.most_common()

  def imprime_diccionario(self):
    print(self.diccionario)
    print(self.diccionarioOrdenado)
```
