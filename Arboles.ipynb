```python
class Nodo():
  #constructor
  def __init__(self, value):
    self.izq = None
    self.dato = value
    self.der = None
    self.padre = None

  #sobrecargamos str
  def __str__(self):
    if self.dato == None:
      pass
    else:
      return '{}'.format(self.dato)

#declaramos el primer nodo y sus hijos
raiz = Nodo('Esto es una prueba')
hijoizq = Nodo('Soy el hijo izquierdo')
hijoder = Nodo('Soy el hijo derecho')

#asignamos los hijos 
raiz.izq = hijoizq
raiz.der = hijoder

#probamos la clase
print(raiz)
print(raiz.izq)
print(raiz.der)
```

#Árbol adivinador

```python
class ArbolAdivinador():
  def __init__(self):
    self.raiz = Nodo("pajaro")

  def si(self, preg):
    resp = str.lower(input(preg))
    return resp[0] == 's'

  def adivina(self):
    ciclo = True
    while ciclo:
      if not self.si("Estas pensando en un animal? "):
        print("Gracias por jugar")
        break
      
      temporal = self.raiz
      

      while temporal.izq != None:
        if self.si(temporal.dato + "? "):
          temporal = temporal.izq

        else:
          temporal = temporal.der

        animal  = temporal.dato
      
        if self.si("Es un" + animal + "?"):
          print("Soy el más grande!")

        else:
          nuevo = input("¿Qué animal era?")
          info = input("¿Qué diferencia a un " + animal + "de un " + nuevo + "? ")
          temporal.dato = info
          temporal.izq = Nodo(animal)
          temporal.der = Nodo(nuevo)
def main():
  adivino = ArbolAdivinador()
  adivino.adivina()

main()
```

```python
class Node:
  def __init__(self, data):
    self.left = None
    self.right = None
    self.data = data
    self.n = None

  def insert(self, data):
    #cuando el data no es None
    if self.data:
      #cuando el nuevo dato es menor, lo ponemos en la izquierda
      if data <self.data:
        #si esta vacio, lo insertamos
        if self.left is None:
          self.left = Node(data)
        #si no, entonces hacermos recursividad
        else:
          self.left.insert(data)
      elif data > self.data:
        if self.right is None:
          self.right = Node(data)
        else:
          self.right.insert(data)

    else:
      self.data = data


  def print_tree(self):
    print(self.data)

  def hay(self):
    if self.n:
      print("Hay algo")
    else:
      print("No hay algo")

raiz = Node(10)
raiz.print_tree()
raiz.insert(11)
raiz.insert(3)
raiz.insert(4)
raiz.insert(15)
```
