##Análisis árboles binarios de búsqueda
1. Eliminar árbol. 

Dado que para eliminar un arbol unicamente se tiene que asignar self.raiz = None, entonces podemos concluir que la complejidad de este algoritmo es constante, pues indpendientemente de la longitud sólo se hará una operación 

$\therefore\, E(n) \in O(1)$

2. Buscar

El peor caso es cuando tenemos un árbol degenerado y además estamos buscando el último elemento, es decir, el mayor

Esto es ya que en dicho caso se tiene una lista ordenada de menor a mayor y estamos usando el algoritmo de busqueda secuencial. Dicho lo anterior, se concluye que tienen el mismo orden de complejidad, es decir: 

$\therefore\, B(n) \in O(n)$

3. Insertar

De la misma manera que en Buscar, el peor caso es cuando tenemos un árbol degenerado y queremos insertar un elemento mayor a todos los anteriores. 

Si $n$ es la altura del árbol, entonces va a hacer $n+1$ operaciones de búsqueda pues va a recorrer los $n + 1$ nodos verificando que el elemento no existe en el árbol. Luego, hace $n+1$ llamadas al método *insertar_nodo*, una por cada nodo en el árbol. Como en cada llamada a *insertar_nodo* hace 4 operaciones (una validación y tres asignaciones), entonces el númeor total de operaciones que realiza es:

$$I(n) = (n + 1) + (n+1)(4) = n + 1 + 4n + 4 = 5n + 5$$
Por lo tanto $I(n) = 5n + 5$
Finalmente, acotando la función:
$$I(n) = 5n + 5 \leq 5n + 5n = 10n$$
Se concluye que *Insetar* pertenece al orden de complejidad lineal
$\therefore\, B(n) \in O(n)$

## Análisis

1. **Eliminar árbol**

  Dado que para eliminar un arbol unicamente se tiene que asignar self.raiz = None, entonces podemos concluir que la complejidad de este algoritmo es constante, pues indpendientemente de la longitud sólo se hará una operación 

$$\therefore\, T(n) \in O(1)$$

2. **Buscar**

  En los árboles balanceados no tenemos peor caso, ya que los datos se encuentran distribuidos de la manera más uniforme posible.

  Para obtener el número de pasos que se realizan para un árbol con $n$ nodos ($T(n)$), consideremos lo siguiente:

  - Supongamos que tenemos el mejor caso, es decir, un árbol completo.

  Si el dato que buscamos es la raíz, sólo se realiza una operación que es la comparación inicial, por lo que $T(1)=a$, con $a$ una constante.

  Si el dato que buscamos es mayor que la raíz, buscamos el dato en el hijo derecho de la raíz (deshaciéndonos así de la otra mitad de los datos en donde ya no se buscará), por lo que se realizarán el número de pasos que se utilizan para buscar en un árbol balanceado de $\frac{n}{2}$ nodos más el número de pasos que se realizaron con la raíz, es decir, $T(n)=T(\frac{n}{2})+a$.

  Si el dato que buscamos es menor que la raíz, buscamos el dato en el hijo izquierdo de la raíz, por lo que se realizarán el número de pasos que se utilizan para buscar en un árbol balanceado de $\frac{n}{2}$ nodos más el número de pasos que se realizaron con la raíz, es decir, $T(n)=T(\frac{n}{2})+a$.

  $$ \therefore\ T(n)= \begin{cases}
a & n=1\\
T(\frac{n}{2})+a & n>1
\end{cases}$$

  **P.D. $T(n) \in O(\log_2 n)$**

  $$ T(n)= \begin{cases}
  a & n=1\\
  T(\frac{n}{2})+a & n>1
\end{cases}$$

  Tomemos $n=2^k$

  \begin{align*}
    T(n)  &=T(2^k)\\
          &=T(\frac{2^k}{2})+a~~~~~Def. ~de~ T(2^k)\\
          &=T(2^{k-1})+a~~~~Leyes~ exponentes\\
          &=(T(2^{k-2})+a)+a~~~~~Def.~ de~ T(2^{k-1})\\
          &=T(2^{k-2})+2\cdot a~~~~~Asociatividad\\
          &=(T(2^{k-3})+a)+2\cdot a~~~~~Def.~ de~ T(2^{k-1})\\
          &=T(2^{k-3})+3\cdot a~~~~~Asociatividad\\
    i-veces & \vdots & \\
           &=T(2^{k-i})+i\cdot a\\
    i=k		& \vdots & \\
            &=T(2^{k-k})+k\cdot a\\
            &=T(2^0)+k\cdot a\\
            &=T(1)+k\cdot a~~~~Leyes~ exponentes\\
            &=a+k\cdot a~~~~Def.~T(1)\\
            &=a+ \log_2 n \cdot a~~~~k=\log_2 n\\
            &=a+ a\log_2 n 
  \end{align*}

  Como a es una constante, la función que crece más rápido es $a \log_2n$, donde si tomamos $c=a$, por la definición de cota superior asintótica, $T(n) \in O(\log_2n) \blacksquare $

3. **Insertar**

  No hay peor caso, ya que los datos se encuentran distribuidos de la manera más uniforme posible.

  Para obtener el número de pasos que se realizan para un árbol con $n$ nodos ($T(n)$), consideremos lo siguiente:

  - Supongamos que tenemos un árbol completo y que el dato que insertaremos no se encuentra en el árbol.

  Primero que nada, se requerirá buscar el nodo, y como no se encuentra en el árbol, esto tomará $\log_2 n$ pasos para realizarse.

  Después, si el dato a insertar es mayor que la raíz,se insertará en el hijo derecho de la raíz (realizando el algoritmo *insertar* para el sub-árbol derecho que tiene $\frac{n}{2}$ nodos), por lo que se realizarán el número de pasos que se utilizan para insertar en un árbol balanceado de $\frac{n}{2}$ nodos, es decir, $E(n)=E(\frac{n}{2})$.

  Si el dato a insertar es mmenor que la raíz,se insertará en el hijo izquierdo de la raíz (realizando el algoritmo *insertar* para el sub-árbol izquierdo que tiene $\frac{n}{2}$ nodos), por lo que se realizarán el número de pasos que se utilizan para insertar en un árbol balanceado de $\frac{n}{2}$ nodos, es decir, $E(n)=E(\frac{n}{2})$.

  Si el sub-árbol donde insertaremos el dato es vacío, el dato se inserta como  hijo de dicho sub-árbol, por lo que $E(1)=a$ con $a$ una constante.

  $$\therefore E(n)= \begin{cases}
  a & n=1\\
  E(\frac{n}{2})+a & n>1
\end{cases}$$

$$\therefore\ T(n)= \log_2n + E(n)$$

**P.D. $T(n) \in O(\log_2 n)$**

  $$E(n)= \begin{cases}
  a & n=1\\
  E(\frac{n}{2})+a & n>1
\end{cases}$$

  Tomemos $n=2^k$

  \begin{align*}
    E(n)  &=E(2^k)\\
          &=E(\frac{2^k}{2})+a~~~~~Def. ~de~ E(2^k)\\
          &=E(2^{k-1})+a~~~~Leyes~ exponentes\\
          &=(E(2^{k-2})+a)+a~~~~~Def.~ de~ E(2^{k-1})\\
          &=E(2^{k-2})+2\cdot a~~~~~Asociatividad\\
          &=(E(2^{k-3})+a)+2\cdot a~~~~~Def.~ de~ E(2^{k-1})\\
          &=E(2^{k-3})+3\cdot a~~~~~Asociatividad\\
    i-veces & \vdots & \\
           &=E(2^{k-i})+i\cdot a\\
    i=k		& \vdots & \\
            &=E(2^{k-k})+k\cdot a\\
            &=E(2^0)+k\cdot a\\
            &=E(1)+k\cdot a~~~~Leyes~ exponentes\\
            &=a+k\cdot a~~~~Def.~E(1)\\
            &=a+ \log_2 n \cdot a~~~~k=\log_2 n\\
            &=a+ a\log_2 n 
  \end{align*}

$$\therefore E(n)=a+ a\log_2 n$$
$$\therefore T(n)=\log_2 n+ a+ a\log_2 n$$
$$\therefore T(n)= a+ (a+1)\log_2 n$$

Como a es una constante, la función que crece más rápido es $(a+1) \log_2n$, donde si tomamos $c=a+1$, por la definición de cota superior asintótica, $T(n) \in O(\log_2n) \blacksquare $

#Árboles de la Barranca
1. Árboles binarios de búsqueda.

Una ventaja es que al hacer el recorrido en orden obtenemos todos los elementos del árbol ordenados en manera ascendente. Una de sus aplicaciones es cuando tenemos muchos datos que entran y salen constantemente

2. Árboles binarios balanceados

Se tiene la misma ventaja que el árbol binario de busqueda además de que el método *buscar* está optimizado pues dicha operación es del orden $O(\log_2(n)$

3. B-Tree

Una ventaja es que siempre está balanceado, pues todas las hojas están al mismo nivel. La operación de eliminar es mucho más eficiente. 

Son usados en SQL para buscar elementos de manera optima 

4. Red-black tree

La complejidad de sus operaciones buscar, eliminar e insertar es del orden $O(\log_2(n))$

Alguno de sus usos es en la geometría computacional

5. AA tree

Tiene timpos de busqueda ligeramente más rápidos que los red-black tree


6. Skiplist

Dado que no hay rotaciones, el método insertar es más rápido, además de que una skiplist es más simple de implementar que un árbol binario o una tabla hash

Apache Portable Runtime usa skiplist


7. Max-min Heap
El orden de complejidad de obtener el elemnto máximo o mínimo en el max y min heap, respectivamente, es constante, pues están localizados en la raíz

Son usados en algoritmos de teoría de gráficas, como el algoritmo Dijkstra

8. Treap
Una ventaja es que son más faciles de coficiar que los árboles de busqueda binaria, pues no necesitant tantos subcasos

Son usados para simular la construcción de un árbol de busqueda binaria aleatorio
9. Scapegoat tree

Aquí no se necesita de memoria adicional para saber si el árbol está balanceado

Tiene las mismas aplicaciones que el árbol bnario de búsqueda
10. Splay tree

Su ventaja es que los nodos a los cuales se accesa con mayor frecuencia van a permanecer cerca de la raíz

De igual manera, se usa frecuentemente cuando tenemos datos que entran y salen constantemente
