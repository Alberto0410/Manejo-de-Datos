#Descripción de los algoritmos

##Búsqueda secuencial
El algoritmo de la búsqueda secuencial recibe como parámetros una lista (la cual puede o no estar ordenada) y un elemento a buscar (de igual manera puede o no estar en la lista dada). Lo que hace es recorrer los elementos de la lista preguntándose si en donde está parado es el elemento que está buscando, si no es así entonces avanza un paso más hasta llegar al elemento buscado o hasta terminar de recorrer la lista, lo que sucede primero. Si lo llega encontrar entonces retorna un True o bien la posición en donde se encuentra. Si no está en la lista entonces regresa un False\
Hay dos peores casos (ambos de la misma magnitud) uno de ellos es cuando el elemento no se encuentra en la lista, ya que el algoritmo va a hacer tantas iteraciones como la longitud del vector. El otro peor caso es cuando el elemento está en al última posición, pues de igual manera va a recorrer toda la lista hasta encontrarlo

##Búsqueda binaria
La búsqueda binaria es un algoritmo para encontrar un elemento dentro de una lista ordenada de elementos. Lo que hace este algoritmo es dividir repetidamente a la mitad la parte de la lista en la que se podría  contener al elemento buscado, hasta reducir a 1 la longitud de dicha parte de la lista . 

Es decir, si tenemos una lista de longitud $n$ y un dato que queremos buscar en dicha lista, si $n = 0$ el algoritmo termina y nos regresa $False$

En otro caso, se toma $n//2$ (la división entera entre en y 2), cuyo valor será denotado como "mitad", si el elemento que se encuentra en la posición "mitad" es el dato que buscamos, el algoritmo termina y se devuelve $True$.

Si  el elemento que se encuentra en la posición "mitad" es menor que el dato que se busca, se vuelve a realizar el algoritmo pero con la lista que empieza desde el elemento en la posición "mitad" hasta el elemento final de ésta.

Si  el elemento que se encuentra en la posición "mitad" es mayor que el dato que se busca, se realiza el algoritmo con la lista que comienza desde el primer elemento de la lista original y termina con el elemento en la posición "mitad".

Si después de que el algoritmo se haya realizado tantas veces hasta que se obtiene una lista de longitud 0, quiere de decir que el elemento no se encontraba en dicha lista y se devuelve $False$.

#Análisis de los algoritmos

##Búsqueda secuencial

###Búsqueda secuencial iterativa
**Análisis tiempo**

Consideremos el código:
```
def busqueda_secuencialI(lista, elemento):
  for i in lista:
    if i == elemento:
      return i
  return False 
```
tenemos que el número de operaciones es:
$$T(n) = \sum_{i=0}^{n-1}(1\, validacion\, + \, 1\, retorno) + 1\, retorno$$
$$T(n) = \sum_{i=0}^{n-1}(2) + 1=2n + 1$$
Por lo tanto: $$T(n) = 2n+1$$



**Análisis espacio**

Al analizar el manejo de memoria notamos que éste es independiente de la longitud de la lista ingresada, siempre hace una única llamada. Dicho esto, no importa si consideramos el peor o el mejor caso, el orden de complejidad de la memoria es constante



###Búsqueda secuencial recursiva
**Análisis tiempo**
Consideremos el código:
```
def busqueda_secuencialR(lista, elemento, indice):
  if len(lista) == indice:
    return False
  if lista[indice] == elemento:
    return indice
  else:
    indice +=1
    return(busqueda_secuencialR(lista, elemento, indice))
```
Tenemos que si $n = 1$, entonces $T(n) = 2$ pues hace una validación y un *return*. 

Por otro lado, si $n\not =1$, entonces:
 $$T(n) = n*(1\, incremento +\, 1\, validacion\, +\, 1\, return)$$

Como estamos considerando el peor caso (cuando no se encuentra en la lista), entonces el número de veces que llama el algoritmo es $n+1$, pues recorre todo los elementos de la lista (por eso la n) y llama una última vez al algoritmo, el cual verifica que el índice es igual a la longitud de la lista y se obtiene $n+1$, por lo tanto:
$$T(n) = n+1$$


**Análisis memoria**

Considerando de nuevo el caso en el que el elemento no se encuentra en la lista, el número de veces que se hace la llamada al algoritmo es $n+2$, pues la primera llamada es la "inicial", después hace $n$ llamadas para recorrer toda la lista y al final hace una última llamada que es cuando se da cuenta que el índice es igual a la longitud de la lista y por lo tanto regresa un False. Así, se tiene que:

$$T(n) = n+2$$


##Búsqueda binaria

###Búsqueda binaria iterativa
**Análisis de tiempo**

Consideremos el código:
```
def busqueda_binariaI(lista, elemento):
  inicio = 0 
  final = len(lista)-1  
  medio = 0 
  bandera = True 
  while bandera:
    if inicio > final:
      return False
    medio = (inicio + final) // 2 
    if lista[medio] < elemento: 
      inicio = medio + 1 
    elif elemento < lista[medio]:
      final = medio - 1
    if lista[medio] == elemento:
      return True
```
Dado el algoritmo, el peor caso en la búsqueda binaria es cuando el elemento no se encuentra en la lista, ya que no habrá momento en que el algoritmo se detenga para regresar *True* o *False* sino hasta que termine de recorrer todas las sublistas de tamaño $\frac{n}{2}$
hasta llegar a la lista de tamaño 0.

Dado el código, tenemos que:

Cuando $n=0$ el algoritmo realiza únicamente una comparación (0$>n-1$=-1) y regresa *False*. Por lo que $T(0)=1 ~paso$

Cuando $n=1$.

$T(1)= T(0)+1~comparación ~+~ 2~operaciones+1~asignación+2~comparaciones+1~operación+1~asignación+1comparación = 10~pasos$

Cuando el tamaño de la lista es $n$ con $n\geq 2$.

$T(n)=T(\frac{n}{2})+ T(1)=T(\frac{n}{2})+ 10$

Entonces:

$$T(n)=\begin{cases}
1 & n=0\\
10 & n=1\\
T(\frac{n}{2})+10& n\geq2
\end{cases} $$

**Análisis de Memoria**

Por la naturaleza del código recursivo, cuando se ejecuta el algoritmo únicamente se hace una llamada a la función, puesto que dentro de ella se ingresa en un ciclo while y todo lo que se modifica en él se va guardando en variables, por lo que no hace algún uso extra de la memoria.


###Búsqueda binaria recursiva
El peor caso en la búsqueda binaria recursiva es cuando el elemento no se encuentra en la lista, entrará al ciclo recursivo tantas veces como este lo permita hasta llegar a alguno de los casos base, que son cuando $n=0$ y cuando $n=1$, donde en ambos casos el número de pasos que se realizan es constante, es decir $T(0)=T(1)=c$ con $c$ una constante.

Ahora, para una lista de tamaño $n$ con $n\geq 2$, por la naturaleza del algoritmo obtenemos que los pasos que realiza el algoritmo es igual al número de pasos que realiza para una lista de $n-1$ elementos más una constante (que se obtiene al llegar a los casos base), por lo que entonces $T(n)= T(\frac{n}{2})+c$ con $c$ una constante.

Entonces:

$$T(n)=\begin{cases}
a & n=0~ó~n=1\\
T(\frac{n}{2})+a& n\geq2
\end{cases} $$

**Análisis de Memoria**

Por el análisis del tiempo, nos damos cuenta que cuando ingresamos una lista de tamaño $n$ se hace una llamada inicial a la función, posteriromente se hace el mismo número de llamadas que las que hacepara la lista de tamaño $\frac{n}{2}$.

Por lo que si $M(n)$ es el número de veces que llama a la funcipon recursiva, tenemos que
$$M(n)=M\left( \frac{n}{2} \right)+1$$

````python
'''
Tarea 2: Algoritmos de búsqueda.

Integrantes:

Martínez Hernpandez Dafne Angélica
Olvera Trejo Alberto

'''
# OTA
#Búsqueda secuencial iterativa
def busqueda_secuencialI(lista:list, elemento:int) -> int:
  '''
  Funcion que realiza el algoritmo de busqueda secuencial de manera iterativa
  Entrada:
        lista: coleccion de datos
        elemento: elemento a buscar en la lista(no necesariamente int)
  Salida: El indice si el elemento se encuentra en la lista
          False si el elemento no se encuentra en la lista

  >>>>>>busqueda_secuencialI([0,2,4,10,7], 4)
  2
  '''
  for i in range(len(lista)): #Recorremos toda la lista
    if lista[i] == elemento: #En cada iteración validamos si el elemento es el buscado
      return i #Si es así, lo retornamos
  return False #En otro caso regresamso un False

# OTA
#Búsqueda secuencial recursiva
def busqueda_secuencialR(lista:list, elemento, indice:int) -> int:
  '''
  Funcion que realiza el algoritmo de busqueda secuencial de manera recursiva
  Entrada:
        lista: coleccion de datos
        elemento: elemento a buscar en la lista(no necesariamente int)
        indice: lugar en el cual se empieza la busqueda
  Salida: El indice si el elemento se encuentra en la lista
          False si el elemento no se encuentra en la lista

  >>>busqueda_secuencialR([0,2,4,10,7], 4, 0)
  2
  '''
  if len(lista) == indice:#Si ya se recorrio toda la lista, regresa un False
    return False 
  if lista[indice] == elemento: #Es el caso base
    return indice
  else: #Si no lo encuentra, entonces hace una llamada recursiva
    return(busqueda_secuencialR(lista, elemento, indice+1))

# MHDA
#Búsqueda binaria iterativa
def busqueda_binariaI(lista:list, elemento) -> int:
  '''
  Funcion que realiza el algoritmo de busqueda de manera iterativa
  Entrada:
        lista: coleccion de datos ordenada
        elemento: elemento a buscar en la lista(no necesariamente int)
  Salida: El indice si el elemento se encuentra en la lista
          False si el elemento no se encuentra en la lista

  >>>busqueda_binariaI([0,2,4,10,7], 4)
  2
  '''

  inicio = 0 # Esta variable indica el primer indice de la lista
  final = len(lista)-1 # Esta variable indica el ultimo indice de la lista
  medio = 0 # Declaramos la variable que se usara en el ciclo while
  bandera = True # Variable de ayuda para la continuacion del ciclo while
  while bandera:
    if inicio > final: # Se evaula si la lista tiene longitud 0
      return False
    medio = (inicio + final) // 2 # Se guarda en una variable el indice del 
                                  # elemento que se encuentra a la mitad de la lista
    if lista[medio] < elemento: # Se evalua si el elemento de el medio se encuentra
                                # en el intervalo izquierdo de la coleccion
      inicio = medio + 1 # Se le da valor al primer indice de la lista
                         # de longitud n-2
    elif elemento < lista[medio]:# Se evalua si el elemento de el medio se encuentra
                                # en el intervalo derecho de la coleccion
      final = medio - 1 # Se le da valor al ultimo indice de la lista
                         # de longitud n-2
    if lista[medio] == elemento: # Se evalua si el elemento que se encuentra en
                                 # el medio es el elemento buscado
      return medio

# MHDA
#Búsqueda binaria recursiva

def busqueda_binariaR(lista:list, elemento, inicio:int, fin:int) -> int:
  '''
  Funcion que realiza el algoritmo de busqueda de manera recursiva
  Entrada:
        lista: coleccion de datos ordenada
        elemento: elemento a buscar en la lista(no necesariamente int)
        inicio: primer indice de la coleccion 
        fin: longitud de la coleccion menos 1 (len(lista)-1)
  Salida: El indice si el elemento se encuentra en la lista
          False si el elemento no se encuentra en la lista
  
  >>>busqueda_binariaR([0,2,4,10,7], 4, 0, len([0,2,4,10,7])-1)
  2
  '''

  if inicio > fin:
    return False
  medio= (inicio + fin)//2 # Se guarda en una variable el indice del 
                           # elemento que se encuentra a la mitad de la lista
  if lista[medio]== elemento: # Se evaula si la lista tiene longitud 0
    return medio
  if lista[medio] < elemento: # Se evalua si el elemento de el medio se encuentra
                                # en el intervalo izquierdo de la coleccion
    return busqueda_binariaR(lista, elemento, medio+1, fin) # Se realiza el algoritmo
        # para la lista de tamaño n-2 que se encuentra del lado izquierdo

  if elemento < lista[medio]:# Se evalua si el elemento de el medio se encuentra
                                # en el intervalo derecho de la coleccion
    return busqueda_binariaR(lista,elemento, inicio, medio-1) # Se realiza el algoritmo
        # para la lista de tamaño n-2 que se encuentra del lado derecho

````

### Código de algoritmos de ordenamiento

````python
#Anexamos los códigos de los algoritmos de ordenamiento de la tarea pasada
def seleccion(lista:list) -> list:
  n= len(lista)
  for i in range(0,n):
    minimo=i
    for j in range(i+1,n):
      if lista[j]<lista[minimo]:
        minimo=j
    lista[i],lista[minimo]=lista[minimo],lista[i]
  return lista

def insertar(lista:list) -> list:
  for elemento in range(len(lista)):
    for anterior in range(elemento,0,-1):
      if lista[anterior]<lista[anterior-1]:
        lista[anterior],lista[anterior-1]=lista[anterior-1],lista[anterior]
  return lista

def burbuja(lista:list) -> list:
    n = len(lista)
    bandera = True
    j = 0
    while(bandera):
        j = j+1
        bandera = False
        for i in range(n-j):
            if lista[i] > lista[i+1] :
                lista[i], lista[i+1] = lista[i+1], lista[i]
                bandera = True
    return lista

def unir(lista:list) -> list:
  if len(lista)<2:
    return lista

  mitad=len(lista)//2
  sublista1=lista[:mitad]
  sublista2=lista[mitad:]
  unir(sublista1)
  unir(sublista2)

  i,j, k = 0,0,0

  while i < len(sublista1) and j< len(sublista2):
    if sublista1[i] < sublista2[j]:
      lista[k]=sublista1[i]
      i+=1
      k +=1
    else:
      lista[k]=sublista2[j]
      j+=1
      k +=1

  while i<len(sublista1):
    lista[k]=sublista1[i]
    i += 1
    k += 1

  while j<len(sublista2):
    lista[k]=sublista2[j]
    j += 1
    k += 1
  
  return lista


def crear_sublistas(lista:list) -> list:
  izquierda = []
  derecha = []
  pivote = lista[0]
  
  for i in range(1, len(lista)):
    if lista[i] < pivote:
      izquierda.append(lista[i])

    else:
      derecha.append(lista[i])
  return izquierda, derecha, pivote


def rapido(lista:list) -> list:
  if 2>len(lista):
    return lista
  else:
    izquierda, derecha, pivote = crear_sublistas(lista)
    return rapido(izquierda) + [pivote] + rapido(derecha)
  
````

````python
# Clase para generar objetos de tipo clientes
class Clientes:

    # Constructor de clientes
    def __init__(self, nombre:str, ID:int, RFC:int):
      '''
      Funcion que realiza el algoritmo de busqueda de manera recursiva
      Entrada:
          Nombre: Cadena de caracteres
          ID: Entero de tamaño 4
          RFC: Entero de tamaño 8 
      '''
      self.nombre = nombre
      self.IDCuenta = ID
      self.RFC = RFC

    # Sobrecarga de los operadores de comparación
    # Determina si un cliente es igual a otro
    def __eq__(self, otro):
        if self.IDCuenta == otro.IDCuenta:
            return True
        else:
            return False

    # Determina si un cliente es mayor o igual a otro
    def __ge__(self, otro):
        if self.IDCuenta >= otro.IDCuenta:
            return True
        else:
            return False

    # Determina si un cliente es menor o igual a otro
    def __le__(self, otro):
        if self.IDCuenta <= otro.IDCuenta:
            return True
        else:
            return False

    # Determina si un cliente es diferente a otro
    def __ne__(self, otro):
        if self.IDCuenta != otro.IDCuenta:
            return True
        else:
            return False

    # Determina si un cliente es mayor a otro
    def __gt__(self, otro):
        if self.IDCuenta > otro.IDCuenta:
            return True
        else:
            return False

    # Determina si un cliente es menor a otro
    def __lt__(self, otro):
        if self.IDCuenta < otro.IDCuenta:
            return True
        else:
            return False

    # Sobrecarga del operador que imprime objetos de tipo cliente
    def __str__(self):
        return "Cliente:{} \n IDCuenta:{} \n RFC:{}\n".format(self.nombre, self.IDCuenta, self.RFC)
````
